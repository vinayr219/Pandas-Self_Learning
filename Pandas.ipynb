# Maps 
The function you pass to map() should expect a single value from the Series (a point value, in the above example), and return a transformed version of that value. 

```python
age_mean=athlete_events.Age.mean()
athlete_events.Age.map(lambda p:p-age_mean)
```

# Apply

call a custom method on each row


```python
def transform_data(row):
    row.Age=row.Age+age_mean
    row.Year=row.Year+10000
    return row

athletes=athlete_events.iloc[:10000,:].apply(transform_data,axis='columns')
```

```python
athlete_events.head(100)
```

```python
athletes
```

# Note :
If we had called reviews.apply() with axis='index', then instead of passing a function to transform each row, we would need to give a function to transform each column. 

```python
def transformAge(column):
    column=column*100
    return column

athletes.Age.apply(transformAge)
```
